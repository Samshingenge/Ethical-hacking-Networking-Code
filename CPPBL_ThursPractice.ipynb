# Working with Strings

```python
#indexing and slicing
#every character is and index
name ="Dr. Hage Geingob"

print(len(name)) #gives the number of characters in a string
#index starts at ) untit size -1
print(name[0])
print(name[len(name)-1])
print(name[7])


#slicing [start,end] buth=t he results exclude the character at end index

print(name[4:7])
print(name[9:]) #prints only from the 9th character

#negative index and slicing

print(name[-1]) #prints last charactor
print(name[-3]) #prints 3rd last character
print(name[-3:])

#invalid index
print(name[16]) #there is nothing at that index so it wont return anything and will give an error
```
